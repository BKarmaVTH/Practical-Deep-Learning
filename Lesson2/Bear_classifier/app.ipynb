```python
from fastai.vision.all import *
import ipywidgets as widgets
import io


learn_inf = load_learner('export.pkl')

btn_upload = widgets.FileUpload()
out_pl = widgets.Output()
lbl_pred = widgets.Label()
btn_run = widgets.Button(description = "Classify")

def on_click_classify(change):
    if not btn_upload.value: return

    img_file = btn_upload.value[0]
    content = img_file.content
    img = PILImage.create(io.BytesIO(content))
    out_pl.clear_output()

    with out_pl: display(img.to_thumb(128,128))

    pred, pred_idx, probs = learn_inf.predict(img)
    lbl_pred.value = f"Prediction: {pred};\t Probability: {probs[pred_idx]:.04f}"

btn_run.on_click(on_click_classify)

widgets.VBox([
    widgets.Label("Select your bear!"),
    btn_upload,
    btn_run,
    out_pl,
    lbl_pred
])
```
