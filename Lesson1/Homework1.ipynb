### Exercise 1

Try to use the dataset URLs.PASCAL_2007 to find the masks

```python
from fastai.vision.all import *
from fastcore.all import *

path1 = untar_data(URLs.PASCAL_2007)
dls1 = SegmentationDataLoaders.from_label_func(
    path1, bs = 8, fnames = get_image_files(path1/"train"),
    label_func = lambda o: path1/'segmentation'/f'{o.stem}_P{o.suffix}',
    codes = ['Background', 'Person', 'Animal', 'Vehicle', 'Object']
)
```

```python
dls1.show_batch(max = 12)
```

```python
learn1= unet_learner(dls1, resnet34)
learn1.fine_tune(3)
```

```python
learn1.show_results()
```

### Exercise 2

Apply the tabular model to following dataset 

URL: https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv

1. Find which chemical parameters influence the **quality** of the wine
2. Write the function **predict_wine(alcohol, ph, acidity)** that returns the quality of the wine
3. Find the wines where the model went the wrong way the most

```python
from fastai.tabular.all import *
from fastai.vision.all import *
import csv


url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"

df = pd.read_csv(url, sep = ";")
print(df.head())
```

```python
df.to_csv("wine.csv", index = False)
```

```python
dls2 = TabularDataLoaders.from_df(
    df,
    cont_names = ["fixed acidity", "volatile acidity", "citric acid", "residual sugar", 
                 "chlorides", "sulphates", "alcohol"],
    cat_names = [],
    y_names = "quality",
    procs = [FillMissing, Normalize]
)

dls2.show_batch()
```

```python
learn2 = tabular_learner(dls2, metrics = rmse) # rmse -> Root Mean Squared Error
learn2.fit_one_cycle(3)
```

```python
learn2.show_results()
```
